# 1D Density Visualization

- also known as a univariate density plot or simply a density plot
- represents the distribution of a single variable along one dimension
- showing how likely different values are along the axis of interest


## Histogram
In a histogram, the data range is divided into bins, and the height of each bin represents the frequency or density of data points within that range.

- Aesthetics used: bar length

```python
# Create a Histogramm from the data
plt.hist(penguins['body_mass_g'])

# Add labels
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogramm')
```

## Kernel density estimate (KDE)
KDE plots provide a continuous estimate of the density.

- Aesthetics used: height of line

```python
sns.kdeplot(data=penguins['body_mass_g'])

# Add labels
plt.xlabel('Wert')
plt.ylabel('Frequency')
plt.title('Kernel density estimate ')
```
