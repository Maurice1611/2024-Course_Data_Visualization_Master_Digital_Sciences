```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('data/city_temperature.csv')
# Filter out years smaller than 1900
data = data[data['Year'] >= 1900]
data
```

```python
# Group by Region and calculate the average temperature
average_temperatures_by_region_year = data.groupby(['Region','Year'])['AvgTemperature'].mean().reset_index()

# Rename the columns
average_temperatures_by_region_year.columns = ['Region', 'Year', 'AvgTemperature']

print(average_temperatures_by_region_year)
```

```python
# Create a pivot table
heatmap_data = average_temperatures_by_region_year.pivot_table(index='Region', columns='Year', values='AvgTemperature')

# Create the heatmap
plt.figure(figsize=(20, 12))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".1f", linewidths=.5)
plt.title('Average Temperature by Region and Year')
plt.xlabel('Year')
plt.ylabel('Region')
plt.show()
```
